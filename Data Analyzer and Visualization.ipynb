```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
city = pd.read_csv('test.csv')
```

```python
#to check the column names and data
city.head()
```

```python
#to look at the data types of the data(per column)
city.info()
```

```python
#data for the air quality in a jointplot graph
sns.jointplot(city['Air_Quality_Index'])
```

```python
#data for health care in a distribution graph
sns.distplot(city['Healthcare_Index'])
```

```python
#The graph below is the representation of multiple factors data  
sns.pairplot(city[['Decibel_Level','Happiness_Score','Healthcare_Index']])
```

```python
#To analyse the traffic density, how often theres traffic density.
sns.stripplot(city['Traffic_Density'])
```

```python
#linear regression graphs comparing the happiness of a person coressponding to thier cost of living
sns.lmplot(x='Happiness_Score', y='Cost_of_Living_Index',data=city)
```
